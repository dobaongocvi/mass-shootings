```python
ms = pd.read_csv('../mass_shootings.csv', index_col = 0)
for i in ms.index:
    if ms.loc[i, 'Decade'] == '70s':
        ms.loc[i, 'Decade'] = '1970s'
    elif ms.loc[i, 'Decade'] == '80s':
        ms.loc[i, 'Decade'] = '1980s'
    elif ms.loc[i, 'Decade'] == '90s':
        ms.loc[i, 'Decade'] = '1990s'
d1 = ms.groupby('Decade')['Incident_ID'].count()
d1.plot(title = 'Mass Shootings by Decade')
d1.to_excel('plot2.xlsx')
```
