# Import Libraries

```python
import pandas as pd
from datetime import *
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Import Data

```python
ms = pd.read_excel('../final data/092622_ALL_mass_schools_distances_REVISED.xlsx').iloc[1:34, 1:21]
alls = pd.read_excel ('../final data/102622_All_Schools_Nearest_TC.xlsx')
ams = pd.read_excel('../final data/Shooting_DataRevised.xlsx')
```

# Distribution of distances to the nearest trauma centers for all schools

```python
sns.histplot(data = alls, x = 'Distance (mi)').set(title = "Distribution of distances for all schools")
plt.axvline(x=alls['Distance (mi)'].mean(),
            color='red')
min_ylim, max_ylim = plt.ylim()
plt.text(alls["Distance (mi)"].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(alls["Distance (mi)"].mean()))
```

# Distribution of distances to the nearest trauma centers for schools with mass shootings

```python
sns.histplot(data = ms, x = 'Distance (mi)').set(title = "Distribution of distances for mass shooting schools")
plt.axvline(x=ms["Distance (mi)"].mean(),
            color='red')
min_ylim, max_ylim = plt.ylim()
plt.text(ms["Distance (mi)"].mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(ms["Distance (mi)"].mean()))
```

The mean distance to the nearest trauma center for schools that experienced mass shootings is 30.4 miles compared to that of 21.9 miles for all schools. This suggests that students attending schools that were affected by shootings, mass shootings in particular are on average at greater risk of not being able to access adequate medical care in the event of being shot. This places a huge burden on the relevant schools and communities to find care and greatly decreases the chance of survival for the affected students.

# Number of shootings in K-12 schools by decades

```python
ams['Decade'] = None
for i in ams.index:
    if 1970 <= ams.loc[i, 'Date'].year < 1980:
        ams.loc[i, 'Decade'] = '1970s'
    elif 1980 <= ams.loc[i, 'Date'].year < 1990:
        ams.loc[i, 'Decade'] = '1980s'
    elif 1990 <= ams.loc[i, 'Date'].year < 2000:
        ams.loc[i, 'Decade'] = '1990s'
    elif 2000 <= ams.loc[i, 'Date'].year < 2010:
        ams.loc[i, 'Decade'] = '2000s'
    elif 2010 <= ams.loc[i, 'Date'].year < 2020:
        ams.loc[i, 'Decade'] = '2010s'
    else:
        ams.loc[i, 'Decade'] = '2020s'
ams.groupby('Decade')['Incident_ID'].count().plot(title = 'Number of Shootings in K-12 Schools by Decade')
plt.ylabel('Frequency of Shootings')
```

The number of shootings in K-12 schools is increasing rapidly in the past few decades, a worrying trend with serious implications for the safety and health of school-age children in the United States.

# Number of mass shootings in K-12 schools by decade

```python
ms['Decade'] = None
for i in ms.index:
    if 1970 <= ms.loc[i, 'Date'].year < 1980:
        ms.loc[i, 'Decade'] = '1970s'
    elif 1980 <= ms.loc[i, 'Date'].year < 1990:
        ms.loc[i, 'Decade'] = '1980s'
    elif 1990 <= ms.loc[i, 'Date'].year < 2000:
        ms.loc[i, 'Decade'] = '1990s'
    elif 2000 <= ms.loc[i, 'Date'].year < 2010:
        ms.loc[i, 'Decade'] = '2000s'
    elif 2010 <= ms.loc[i, 'Date'].year < 2020:
        ms.loc[i, 'Decade'] = '2010s'
    else:
        ms.loc[i, 'Decade'] = '2020s'
ms.groupby('Decade')['School'].count().plot(title = 'Number of Mass Shootings in K-12 Schools by Decade')
plt.ylabel('Frequency of Shootings')
```

As we can see, the number of mass shootings are concentrated in the most recent three decades. There were no mass shootings in the decade before the 1990s. This is in line with the trend we observed with the frequency of shootings overall in K-12 schools, which has been increasing over the past 5 decades.
